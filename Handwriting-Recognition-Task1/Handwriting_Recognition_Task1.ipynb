## Phase 2: Data Preprocessing
### Normalizing and Reshaping Data

````python
# Before preprocessing
print("Before preprocessing:")
print(f"x_train shape: {x_train.shape}")
print(f"x_train min/max: {x_train.min()}/{x_train.max()}")

# 1. Normalize pixel values from 0-255 to 0-1
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# 2. Reshape for CNN (add channel dimension)
x_train = x_train.reshape(-1, 28, 28, 1)  # -1 means "automatically calculate this dimension"
x_test = x_test.reshape(-1, 28, 28, 1)

print("\nAfter preprocessing:")
print(f"x_train shape: {x_train.shape}")
print(f"x_test shape: {x_test.shape}")
print(f"x_train min/max: {x_train.min()}/{x_train.max()}")

# Let's verify our reshaping worked
print(f"\nSample image shape after reshaping: {x_train[0].shape}")
````

## Phase 3: Model Building
### Creating a Convolutional Neural Network

````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Build the model
model = Sequential([
    # First convolutional layer
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D((2, 2)),

    # Second convolutional layer
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),

    # Flatten and dense layers
    Flatten(),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')  # 10 units for 10 digits (0-9)
])

# Let's see our model architecture
print("Model Architecture:")
model.summary()
````

````python
# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',  # Use this when labels are integers
    metrics=['accuracy']
)

print("Model compiled successfully!")
print(f"Optimizer: {model.optimizer.get_config()['name']}")
print(f"Loss function: {model.loss}")
print(f"Metrics: {model.metrics_names}")
````

## Phase 4: Model Training
### Training and Validating the CNN

````python
# Train the model
print("Starting training...")
history = model.fit(
    x_train,
    y_train,
    batch_size=128,           # Process 128 images at a time
    epochs=5,                 # Number of complete passes through the dataset
    validation_data=(x_test, y_test),  # Check performance on test set
    verbose=1                 # Show progress bars
)

print("\nTraining completed!")
````

## Phase 6: Model Improvements
### Adding Dropout and More Layers

````python
from tensorflow.keras.layers import Dropout, BatchNormalization
from tensorflow.keras.regularizers import l2

# Build an improved model with dropout and more layers
improved_model = Sequential([
    # First convolutional block
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    BatchNormalization(),
    Conv2D(32, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Dropout(0.25),  # Drop 25% of neurons

    # Second convolutional block
    Conv2D(64, (3, 3), activation='relu'),
    BatchNormalization(),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Dropout(0.25),

    # Third convolutional block
    Conv2D(128, (3, 3), activation='relu'),
    BatchNormalization(),
    Dropout(0.25),

    # Classifier
    Flatten(),
    Dense(256, activation='relu', kernel_regularizer=l2(0.001)),
    BatchNormalization(),
    Dropout(0.5),  # Drop 50% of neurons in dense layer
    Dense(128, activation='relu', kernel_regularizer=l2(0.001)),
    Dropout(0.5),
    Dense(10, activation='softmax')
])

# Compile the improved model
improved_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("Improved model architecture:")
improved_model.summary()
````

````python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Create data augmentation generator
# This creates modified versions of our training images
datagen = ImageDataGenerator(
    rotation_range=10,      # Randomly rotate images by ±10 degrees
    zoom_range=0.1,         # Randomly zoom in/out by 10%
    width_shift_range=0.1,  # Randomly shift images horizontally by 10%
    height_shift_range=0.1, # Randomly shift images vertically by 10%
    shear_range=0.1,        # Shear transformations
    fill_mode='nearest'     # How to fill newly created pixels
)

# Fit the data generator to our training data
datagen.fit(x_train)

print("Data augmentation generator created!")
print("Augmentation parameters:")
print(f"Rotation range: ±10 degrees")
print(f"Zoom range: 10%")
print(f"Shift range: 10%")
````

````python
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Define callbacks for better training
callbacks = [
    EarlyStopping(
        monitor='val_loss',     # Monitor validation loss
        patience=5,            # Stop if no improvement for 5 epochs
        restore_best_weights=True  # Keep the best weights
    ),
    ReduceLROnPlateau(
        monitor='val_loss',    # Monitor validation loss
        factor=0.2,           # Reduce learning rate by 80%
        patience=3,           # Wait 3 epochs before reducing
        min_lr=0.00001        # Minimum learning rate
    )
]

print("Starting training with improved model and data augmentation...")

# Train with data augmentation
history_improved = improved_model.fit(
    datagen.flow(x_train, y_train, batch_size=128),  # Use augmented data
    epochs=15,                    # Train for more epochs
    validation_data=(x_test, y_test),
    callbacks=callbacks,          # Use our callbacks
    verbose=1
)

print("\nImproved training completed!")
````
